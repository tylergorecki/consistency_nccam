# Rolling average initial test

My first goal with this project was to understand how to calculate the rolling average over time given a ChatGPT-generated dataset of five possible variables I'd see in a basketball stats dataset for one player over a single season. 

```python
import pandas as pd
import random
from sklearn.preprocessing import MinMaxScaler

# Create a dictionary with sample data
data = {
    'points': [random.randint(5, 30) for _ in range(30)],
    'assists': [random.randint(1, 10) for _ in range(30)],
    'rebounds': [random.randint(3, 15) for _ in range(30)],
    'blocks': [random.randint(0, 5) for _ in range(30)],
    'steals': [random.randint(0, 5) for _ in range(30)]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Initialize MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit and transform the DataFrame
scaled_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
#players = df.iloc[:,0]
#scaled_df.set_index(players, inplace = True)
scaled_df = scaled_df.round(2)

# Calculate rolling sd of the last five rows
rolling_sd = scaled_df.rolling(window=5).std()

# create new cols
scaled_df['rolling_points_sd'] = rolling_sd['points']
scaled_df['rolling_assists_sd'] = rolling_sd['assists']
scaled_df['rolling_rebounds_sd'] = rolling_sd['rebounds']
scaled_df['rolling_blocks_sd'] = rolling_sd['blocks']
scaled_df['rolling_steals_sd'] = rolling_sd['steals']

scaled_df['consistency_last5'] = scaled_df.iloc[:, -5:].mean(axis=1)

print(scaled_df)
```
